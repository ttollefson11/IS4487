## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

### Additional step 1 for cleaning/filtering data
Drop the id column from the df.clean dataframe

```python
df_clean = df_clean.drop(columns = ['id'])
```

### Additional step 2 for cleaning/filtering data
Filter out rows where leftover is less than 10

```python
df_clean = df_clean[(df_clean['leftover'] >= 10)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Slice the data  
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2) 
```

### Additional chart #1
Average income level compared to the customer leaving or staying

```python
df_clean[['income','leave']].groupby('leave').mean('income').plot.bar(color = 'orange')
```

### Additional chart #2
Count of reported satisfaction compared to the customer leaving or staying

```python
df_clean[['reported_satisfaction','leave']].groupby('leave').count().plot.bar(color = 'orange')
```
