# Clean the Data

Remove rows with NAs

```python
df_clean = df.dropna()

df_clean.describe()
```

Remove the row with the single outlier in num_accts. Filter with a logical condition

```python
df_clean = df_clean[(df_clean['num_accts'] < 20)]

df_clean.describe()
```

Remove the rows with negative income values

```python
df_clean = df_clean[(df_clean['income'] >= 0)]

df_clean.describe()
```

Make the 4 variables you have identified in the previous question into factors

```python
df_jobs = pd.get_dummies(df_clean['job'])
print(df_jobs)
```

```python
df_chk_acct = pd.get_dummies(df_clean['chk_acct'])
print(df_chk_acct)
```

```python
df_sav_acct = pd.get_dummies(df_clean['sav_acct'])
print(df_sav_acct)
```

```python
df_product = pd.get_dummies(df_clean['product'])
print(df_product)
```
