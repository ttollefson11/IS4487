## Clean up the data
Create new dataframe with filtered columns that pertain to analysis

```python
df_filtered = df[['created_at', 'text', 'quote_count', 'reply_count', 'retweet_count', 'favorite_count', 'entities', 'favorited', 'retweeted']]
```

Filter the text column in the df_filtered dataset to only include tweets with the text "SuperBowl" and "mmschocolate"

```python
df_filtered_tweets = df_filtered[df_filtered["text"].str.contains("SuperBowl", "mmschocolate")]
```

## Visualize the Data

```python
df_filtered_tweets.plot.scatter(x= "favorite_count", y= "retweet_count")
```

```python
df_filtered_tweets[['created_at','favorite_count']].groupby('favorite_count').count().plot.bar()
```
