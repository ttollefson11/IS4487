## Clean up the data


Create a copy of variable df dataframe

```python
df_copy = df
df_copy.info()
```

Convert the Total_Amount into a number (not a string)

```python
df_copy['total_amount'] = df_copy['total_amount'].str.replace(',', '')
df_copy['total_amount'] = df_copy['total_amount'].str.replace('$', '')
df_copy['total_amount'] = df_copy['total_amount'].astype('float')
df_copy.info()
```

Convert the product_price into a number (not a string)

```python
df_copy['product_price'] = df_copy['product_price'].str.replace(',', '')
df_copy['product_price'] = df_copy['product_price'].str.replace('$', '')
df_copy['product_price'] = df_copy['product_price'].astype('float')
df_copy.info()
```

Remove rows with empty (NULL) values

```python
df_copy = df_copy.dropna(axis = 'index', how = 'any', subset = ['product_id', 'product_name', 'product_price', 'total_amount', 'order_year_month', 'order_month'])
```
